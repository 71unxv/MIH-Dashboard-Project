 # Read and merge between labeled and timeseries data

```python
###########
#define the input filename location
Realtime_DB_filename_list = [
    "sorik/143-1.csv",
    "sorik/143-2.csv",
    "sorik/143-3.csv",
    "sorik/143-4.csv",
    "sorik/143-5.csv",
    "sorik/143-6.csv",
    "sorik/143-7.csv",
    "sorik/143-8.csv",
]
Label_Raw_DB_filename = 'CompileLabel_forPython.csv'

######
#read and load the csv files
# Label_Raw_DB = pd.read_excel(Label_Raw_DB_filename,engine="openpyxl", sheet_name="Raw Data")
Label_Raw_DB = pd.read_csv(Label_Raw_DB_filename,delimiter=";")
# create a DateTime Column
Label_Raw_DB['Date/Time'] = Label_Raw_DB['Date'] + " " + Label_Raw_DB['Time']
Label_Raw_DB['Connection Activity'].fillna('No Data',inplace=True)

# merge several csv files
Realtime_DB_MERGE = pd.DataFrame()
for filename in Realtime_DB_filename_list:
    Realtime_DB_MERGE = Realtime_DB_MERGE.append(pd.read_csv(filename,delimiter=";"))

######
# makesure the timeseries data is well sorted based data&timetime(oldest to newest)
Realtime_DB_MERGE = Realtime_DB_MERGE.sort_values(by = ['date', 'time'], ascending = [True, True], na_position = 'first')

#####
#displaying the loaded data to make sure the csv data are successfully loaded and merged

print("RealTime Database on " + str(Realtime_DB_filename_list))
displayDF(Realtime_DB_MERGE)

print("LabelRaw Database on " + Label_Raw_DB_filename)
displayDF(Label_Raw_DB)
```

```python
print("start realtime:")
print(Realtime_DB_MERGE['dt'].min())
print("\n")
print("stop realtime:")
print(Realtime_DB_MERGE['dt'].max())
print("\n")
print("=="*10)
print("start RTO_labelling:")
print(Label_Raw_DB['Date/Time'].min())
print("\n")
print("stop RTO_labelling:")
print(Label_Raw_DB['Date/Time'].max())
```

 ## merge and label the timeseries data with the existing label_raw data.
 

```python
pd.set_option('display.max_rows', 200)
displayDF(Realtime_DB_MERGE.describe().T)
```

```python
Realtime_DB_MERGE_CLEAN = Realtime_DB_MERGE.loc[:, (Realtime_DB_MERGE != 0.0).any(axis=0)]
displayDF(Realtime_DB_MERGE_CLEAN)
Realtime_DB_MERGE_CLEAN.describe()
```

```python
Realtime_DB_LABELED = Realtime_DB_MERGE.copy()
Realtime_DB_LABELED['LABEL_ConnectionActivity'] = np.nan
Realtime_DB_LABELED['LABEL_SubActivity'] = np.nan
Realtime_DB_LABELED['LABEL_Activity'] = np.nan

for index, Label_Raw_Row in Label_Raw_DB.iterrows():
    # displayDF(Label_Raw_Row)
    # print(Label_Raw_Row['Date/Time'])
    # Connection Activity	Sub Activity	Activity
    StartTime = str(Label_Raw_Row['Date/Time'])
    EndTime = (Label_Raw_DB.loc[index+1,['Date/Time']])
    EndTime = str(EndTime[0])
    # if Label_Raw_Row['Connection Activity']:
    #     LABEL_ConnectionActivity = np.nan
    # else:
    LABEL_ConnectionActivity = str(Label_Raw_Row['Connection Activity'])
    LABEL_SubActivity = str(Label_Raw_Row['Sub Activity'])
    LABEL_Activity = str(Label_Raw_Row['Activity'])
    
    print(str(StartTime) + "  --  "+str(EndTime))
    print(LABEL_ConnectionActivity + "  --  " + LABEL_SubActivity+ "  --  " +LABEL_Activity)
    print("Progress:  " + str(index*100/max(Label_Raw_DB.index.values)))

    row_logic = (Realtime_DB_LABELED['dt']> StartTime) & (Realtime_DB_LABELED['dt']<= EndTime)

    Realtime_DB_LABELED.loc[row_logic,'LABEL_ConnectionActivity'] = LABEL_ConnectionActivity
    Realtime_DB_LABELED.loc[row_logic,'LABEL_SubActivity'] = LABEL_SubActivity
    Realtime_DB_LABELED.loc[row_logic,'LABEL_Activity'] = LABEL_Activity

    # displayDF(Realtime_DB_LABELED.loc[row_logic,:])
    if index+1== max(Label_Raw_DB.index.values):
        break
    # if index== 5:
    #     break


# displayDF(Realtime_DB_LABELED.describe().T)
    
# Temp_DB_date
# Temp_DB_hour
```

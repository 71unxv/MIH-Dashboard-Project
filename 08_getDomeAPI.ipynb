```python
import pandas as pd 
import requests
import json
from IPython.display import display as display_DF
```

```python
JSON_Temp = requests.get(
        "https://pdumitradome.id/dome_api/rtdc/get_company/"
    ).json()
# print(JSON_Temp)
pd.json_normalize(JSON_Temp, record_path = 'result')
```

```python
CompList = json.loads(
    requests.get(
        "https://pdumitradome.id/dome_api/rtdc/get_company/"
    ).text
)

print(CompList)
CompList_DF = pd.json_normalize(CompList, record_path = 'result')
display_DF(CompList_DF)
```

```python
cid = 9
getWellAPI = "https://pdumitradome.id/dome_api/rtdc/get_well?cid=" + str(cid)


WellList = json.loads(
    requests.get(
        getWellAPI
    ).text
)

# print(WellList)

WellList_DF = pd.json_normalize(WellList, record_path = 'result')
WellList_DF['active_date'] = WellList_DF['active_date'].astype('datetime64')
WellList_DF['end_date'] = WellList_DF['end_date'].astype('datetime64')
# WellList_DF['end_date']


display_DF(WellList_DF)
print(WellList_DF.dtypes)
# print(WellList_DF['time_master'].to_list())
```

```python
getData_params ={
	"wid" : 144,
	"start" : "2021-06-03 00:00:00",
   	"end" : "2021-06-05 00:00:00"
}


WellData = (
	(
        requests.get("https://pdumitradome.id/dome_api/rtdc/get_data", data=json.dumps(getData_params))
    ).text
)
# print(json.dumps(WellData))
Data_DF = pd.json_normalize(json.loads(WellData), record_path='result')
display_DF(Data_DF)
```

```python
import pandas as pd
```

```python
import glob

list_WITS_file = glob.glob("Data\\raw_wits\\*.csv")



end_time = []
start_time = []
filename = []
TotalRow = []
TotalColumns = [] 

for i in list_WITS_file:
    df_temp = pd.read_csv(i, delimiter='\t')
    # print(df_temp.columns)
    # print((df_temp.reset_index().iloc[:,1].max()))
    # print((df_temp.reset_index().iloc[:,1].min()))
    # print((df_temp.describe().loc[['min', 'max'],[0]]))
    # print(i.split("\\")[-1])
    # TotalRow.append(df_temp.shape)
    # print(df_temp.shape)
    TotalRow.append(df_temp.shape[0])
    TotalColumns.append(df_temp.shape[1])

    end_time.append((df_temp.reset_index().iloc[:,1].max()))
    start_time.append((df_temp.reset_index().iloc[:,1].min()))
    filename.append(i.split("\\")[-1])


    # break
```

```python
dict_temp = {
    'filename'  :filename,
    'end_time' : end_time,
    'start_time' : start_time,
    'TotalRow':TotalRow,
    'TotalColumns':TotalColumns
}
# end_time = []
# start_time = []
# filename = []
    
df_out = pd.DataFrame(dict_temp)
df_out
```

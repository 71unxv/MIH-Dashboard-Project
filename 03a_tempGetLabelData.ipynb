```python
import pandas as pd 

# FolderPath = "/content/drive/MyDrive/Colab Notebooks/PDU"
FilePath = "OLD\\FinalLabelDataAll.csv"
# /content/drive/MyDrive/Colab Notebooks/PDU/FinalLabelDataAll.csv
Raw_DB = pd.read_csv(FilePath)
Raw_DB.head()
```

```python
# for i in Raw_DB.columns:
#     print(i)

TimebasedCol = [
    "dt",
    "date",
    "time"
]
ListUsedCol = ["bitdepth",
            "md",
            "blockpos",
            "rop",
            "hklda",
            "woba",
            "torqa",
            "rpm",
            "stppress",
            "mudflowin"]

LabelCol = ["LABEL_ConnectionActivity",
            "LABEL_SubActivity",
            "LABEL_Activity"]
        
finalUsedCol = TimebasedCol + ListUsedCol + LabelCol
finalUsedCol
```

```python
Used_DB = Raw_DB[finalUsedCol]
Used_DB[LabelCol[0]] = Used_DB[LabelCol[0]].str.upper()
Used_DB[LabelCol[1]] = Used_DB[LabelCol[1]].str.upper()
Used_DB[LabelCol[2]] = Used_DB[LabelCol[2]].str.upper()

Used_DB["LABEL_SubActivity"] = Used_DB["LABEL_SubActivity"].str.replace('REAMING ', 'REAMING', regex=False)

Used_DB
```

```python
start_time = "4/26/2021 12:56:00"
end_time = "23/05/2021 00:55"

start_time = "2021-05-11 03:10:00"
end_time = "2021-05-23 00:55:00"



idx_logic = (Used_DB["dt"] >= start_time) &  (Used_DB["dt"] <= end_time)
FinalUsed_DB = Used_DB.loc[idx_logic, :]
FinalUsed_DB
```

```python
# Label the drill vs trip data
list_start_time = []
list_end_time = []


list_start_time.append("2021-05-12 15:03:40")
list_end_time.append("2021-05-15 2:41:33")
list_start_time.append("2021-05-15 19:29:28")
list_end_time.append("2021-05-17 17:26:01")
list_start_time.append("2021-05-17 21:53:01")
list_end_time.append("2021-05-19 21:37:21")


FinalUsed_DB.loc[:, "LABEL_MajorActivity"] = "Tripping"

for start_temp,end_temp in zip(list_start_time, list_end_time):

    idx_logic = (Used_DB["dt"] >= start_temp) &  (Used_DB["dt"] <= end_temp)
    FinalUsed_DB.loc[idx_logic, "LABEL_MajorActivity"] = "Drilling"
    print(start_temp)
    print(end_temp)

```

```python
import pandas as pd 

# FolderPath = "/content/drive/MyDrive/Colab Notebooks/PDU"
FilePath = "OLD\\FinalLabelDataAll.csv"
# /content/drive/MyDrive/Colab Notebooks/PDU/FinalLabelDataAll.csv
Raw_DB = pd.read_csv(FilePath)
Raw_DB.head()
```

```python
# for i in Raw_DB.columns:
#     print(i)

TimebasedCol = [
    "dt",
    "date",
    "time"
]
ListUsedCol = ["bitdepth",
            "md",
            "blockpos",
            "rop",
            "hklda",
            "woba",
            "torqa",
            "rpm",
            "stppress",
            "mudflowin"]

LabelCol = ["LABEL_ConnectionActivity",
            "LABEL_SubActivity",
            "LABEL_Activity"]
        
finalUsedCol = TimebasedCol + ListUsedCol + LabelCol
finalUsedCol
```

```python
Used_DB = Raw_DB[finalUsedCol]
Used_DB[LabelCol[0]] = Used_DB[LabelCol[0]].str.upper()
Used_DB[LabelCol[1]] = Used_DB[LabelCol[1]].str.upper()
Used_DB[LabelCol[2]] = Used_DB[LabelCol[2]].str.upper()

Used_DB["LABEL_SubActivity"] = Used_DB["LABEL_SubActivity"].str.replace('REAMING ', 'REAMING', regex=False)

Used_DB
```
